```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
price_df = pd.read_csv("price.csv")

temperature_df = pd.read_csv("merged.csv")
```

```python
price_df['Price Date'] = pd.to_datetime(price_df['Price Date'],format="%d/%m/%Y")
temperature_df['datetime'] = pd.to_datetime(temperature_df['datetime'],format="%d/%m/%Y")
```

```python
merged_df = pd.merge(price_df, temperature_df, left_on=['Market Name', 'Price Date'], right_on=['name', 'datetime'], how='left')
```

```python
merged_df.drop(columns=['District Name','Variety','Sl no.','name','Price Date'], inplace=True)
```

```python
merged_df.drop(columns=['Market Name','Commodity','Grade','datetime'
], inplace=True)
```

```python
merged_df=merged_df[merged_df['Min Price (Rs./Quintal)'].notna()]
```

```python
grouped1 = merged_df.groupby(['tempmax'])
grouped2 = merged_df.groupby(['tempmin'])
grouped3 = merged_df.groupby(['temp'])
grouped4 = merged_df.groupby(['dew'])
grouped5 = merged_df.groupby(['humidity'])
grouped6 = merged_df.groupby(['precip'])
grouped7 = merged_df.groupby(['windspeed'])

```

```python
mean_values1 = grouped1.mean()
mean_values2 = grouped2.mean()
mean_values3 = grouped3.mean()
mean_values4 = grouped4.mean()
mean_values5 = grouped5.mean()
mean_values6 = grouped6.mean()
mean_values7 = grouped7.mean()
```

```python
merged_df['tempmin'] = grouped1['tempmin'].transform(lambda x: x.fillna(x.mean()))
merged_df['tempmax'] = grouped2['tempmax'].transform(lambda x: x.fillna(x.mean()))
merged_df['temp'] = grouped3['temp'].transform(lambda x: x.fillna(x.mean()))
merged_df['dew'] = grouped4['dew'].transform(lambda x: x.fillna(x.mean()))
merged_df['humidity'] = grouped5['humidity'].transform(lambda x: x.fillna(x.mean()))
merged_df['precip'] = grouped6['precip'].transform(lambda x: x.fillna(x.mean()))
merged_df['windspeed'] = grouped7['windspeed'].transform(lambda x: x.fillna(x.mean()))
```
